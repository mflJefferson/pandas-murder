```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('fivethirtyeight')
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
path = 'datasets/'
startName = 'homicidios_de_'
filenames = {'janeiro', 'fevereiro', 'marco', 'abril', 'maio', 'junho', 'julho', 'agosto', 'setembro'}
extension = '.xlsx'
```

```python
def populaDataFrame():
    dfList = list()

    for file in filenames:
        df= pd.DataFrame(pd.read_excel(path+startName+file+extension))
        dfList.append(df)

    df = pd.concat(dfList)
    return df
```

```python
df = populaDataFrame()
```

## 1. Número de homicídios por cidade

```python
df.groupby('CIDADE').size().sort_values().tail(12).plot(kind='barh', figsize=(10,6))
```

## 2. Comparação de vítimas por gênero

```python
df.groupby('SEXO').size().sort_values().tail().plot(kind='pie',title = 'Gênero', figsize=(10,6))
```

## 3. Periodo da ocorrência mais comum em São José dos Campos

```python
df.query('CIDADE == "S.JOSE DOS CAMPOS"').groupby('PERIDOOCORRENCIA').size().sort_values().tail().plot(kind='barh', figsize=(10,6));
```

## 4. Periodo da ocorrência mais comum em Taubaté

```python
df.query('CIDADE == "TAUBATE"').groupby('PERIDOOCORRENCIA').size().sort_values().tail().plot(kind='barh', figsize=(10,6));
```

## 5. Localização mais comuns dos homicídios

```python
df.groupby('DESCRICAOLOCAL').size().sort_values().tail(8).plot(kind='barh', figsize=(10,6))
```
