```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('fivethirtyeight')
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

## dataset

#### Os dados foram consultados desse link ~> http://www.ssp.sp.gov.br/transparenciassp/

```python
path = 'datasets/'
startName = 'homicidios_de_'
filenames = {'janeiro', 'fevereiro', 'marco', 'abril', 'maio', 'junho', 'julho', 'agosto', 'setembro'}
extension = '.xlsx'
```

```python
def populaDataFrame():
    dfList = list()

    for file in filenames:
        df= pd.DataFrame(pd.read_excel(path+startName+file+extension))
        dfList.append(df)

    df = pd.concat(dfList)
    return df
```

```python
df = populaDataFrame()
```

```python
df.groupby('CIDADE').size().sort_values().tail(10).plot(kind='barh', figsize=(10,6))
```

```python
df.groupby('SEXO').size().sort_values().tail(10).plot(kind='pie',title = 'Sexo', figsize=(10,6))
```

```python
df.groupby('NATUREZAVINCULADA').size().sort_values().tail(10).plot(kind='barh', figsize=(10,6))
```
