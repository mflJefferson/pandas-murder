```python
path = 'datasets/'
startName = 'dataset_of_'
filenames = {'janeiro', 'fevereiro', 'marco', 'abril', 'maio', 'junho', 'julho', 'agosto', 'setembro'}
extension = '.xlsx'
```

```python
#Função para percorrer os arquivos de dados e popular o dataframe.
def populaDataFrame():
    dfList = list()

    for file in filenames:
        df= pd.DataFrame(pd.read_excel(path+startName+file+extension))
        dfList.append(df)

    df = pd.concat(dfList)
    return df
```

```python
df = populaDataFrame()
```
